```python
import pandas as pd
data = {
    "IdProducto": [1001, 1002, 1003, 1003],
    "CantidadProducto": [30, None, 25, 25],
    "Precio": [20.5, 15.0, None, 22.5]}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.isnull() #Donde esta True es porque hay un valor nulo
```

```python
df.notnull() #Donde estan los False hay un valor nulo
```

```python
df.isnull().mean()
```

```python
df.notnull().sum()
```

```python
#SOLUCION A LOS NULOS... NO ELIMINARLOS... QUIZA REEMPLAZARLOS DEPENDE...
# dropna.... valores no habilitados
df_eliminados = df.dropna()
df_eliminados
```

```python
valores_nuevos = {"CantidadProducto":0, "Precio":df["Precio"].mean()}
df_rellenados = df.fillna(valores_nuevos)
df_rellenados
```

```python
#modificar datos de una serie de dataframe
df_rellenados["CantidadProducto"] = df_rellenados["CantidadProducto"].astype(int)
df_rellenados
```

```python
#REVISAR SI HAY DUPLICADOS... sus parametros busca que sean duplicados completos
df_unicos = df_rellenados.drop_duplicates()
df_unicos
```

```python
#REVISAR SI HAY DUPLICADOS
df_unicos = df_rellenados.drop_duplicates(subset="IdProducto")
df_unicos
```
