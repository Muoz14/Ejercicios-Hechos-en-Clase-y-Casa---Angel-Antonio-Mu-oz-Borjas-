```python
import numpy as np
import pandas as pd
```

1. Creación del DataFrame a partir de los datos de medallas.csv

```python
ruta = "C:/medallas.csv"
df = pd.read_csv(ruta)

df
```

```python
2. Observaciones de los datos usando Pandas.
```

```python
df.info() #Aqui podemos ver como en las 3 tipos de medallas existen campos nulos
#Y descartamos que el total y el pais contengan datos que son nulos
```

```python
df.head()
```

```python
df.isnull() #Aqui visualizamos de una diferente manera los datos nulos, donde hay un True hay un nulo
```

```python
df.isnull().sum() #Sumamos la cantidad de nulos que hay en cada medalla
```

```python
#Aqui vamos a hacer el cambio de valor para los nulos, rellenaremos esos espacios vacíos específicamente con el   
#numero cero. Si un país tiene un dato faltante en la columna "Bronce" por ejemplo, significa que no ganó ninguna.
valores_nuevos = {"Oro":0, 
                  "Plata":0,
                  "Bronce":0
                 }
df_rellenados = df.fillna(valores_nuevos)
df = df_rellenados

#Y mostraremos el resultado final de nuestro dataframe modificado
df 
```

```python
#Aqui estoy haciendo el cambio de tipo de dato de cada una de las medallas a entero, porque no se puede
#tener 1.5 medallas.
df["Oro"] = df["Oro"].astype(int)
df["Plata"] = df["Plata"].astype(int)
df["Bronce"] = df["Bronce"].astype(int)
df
```

```python
3. Top 3 de paises con mas medallas olimpicas
```

```python
#df.sort_values(by='NombreDeLaColumna', ascending=False) sintaxis de la funcion utilizada
#by se usa para pasar el nombre de la columna (como string) a usar como referencia para ordenar. 
#ascending se usa para recibir un valor booleano (True o False). Por defecto viene en True (ascendente, es decir, 
#del número más pequeño al más grande).
df.sort_values(by='Total', ascending = False).head(3)
```

4. Los 10 paíases que tienen más de 10 medallas olimpicas. 

```python
#Sintaxis DataFrame[ DataFrame['NombreDeLaColumna'] > ValorNumerico ]
df[df['Total'] > 10]
```
