```python
import pandas as pd
from array import array
```

```python
#Arreglo
edades = array("i", [18, 19, 20, 18])
```

```python
#Lista
estudiantes = ["Carlos", "Ana", "Luis", "Maria"]
```

```python
#Diccionario
notas = {
    "Carlos": 85,
    "Ana": 92,
    "Luis": 78,
    "Maria": 90
}
```

```python
print(notas["Ana"])

print("Lista de Estudiantes")
print(estudiantes)

print("\nDiccionario de notas")
print(notas)

print("\nArreglo de Edades")
print(edades)
```

```python
print("\nRecorrido de Datos")

for i in range(len(estudiantes)):
    nombre = estudiantes[i]
    nota = notas[nombre]
    edad = edades[i]

    print(f"Estudiante: {nombre} | Edad: {edad} | Nota: {nota}")
```

```python
print("\nAGREGANDO NUEVO ESTUDIANTE")    

estudiantes.append("Pedro")
notas["Pedro"] = 88
edades.append(21)

for i in range(len(estudiantes)):
    nombre = estudiantes[i]
    nota = notas[nombre]
    edad = edades[i]

    print(f"Estudiante: {nombre} | Edad: {edad} | Nota: {nota}")

pandas_df = pd.DataFrame({
    "Nota": notas.values()
})
```

```python
print("\nDATAFRAME SOLO CON NOTAS")
print(pandas_df)
```
